```python
#this can be run over and over until you're happy with the distribution ;)

control_before = [int(n) for n in np.random.normal(115, 15, 60)]
test_before = [int(n) for n in np.random.normal(115, 15, 40)]
```

```python
plt.hist(control_before, alpha=1.0, normed=False)
plt.show()
plt.hist(test_before, alpha=1.0, normed=False)
plt.show()
```

## Boxplots


```python
# basic boxplot

my_list5 = np.concatenate((np.random.uniform(0.5, 10, 50), np.random.normal(7, 2, 20)))
plt.boxplot(my_list5, vert=False)
plt.show()
```

```python
# basic boxplot with outliers

my_list6 = np.concatenate((np.random.uniform(0.5, 10, 50), np.random.normal(7, 10, 20)))
plt.boxplot(my_list6, vert=False)
plt.show()
```

```python
# basic boxplot with full whiskers

plt.boxplot(my_list6, vert=False, whis='range')
plt.show()
```

```python
# one-dimensional scatterplot

plt.scatter(my_list6, np.random.normal(0, 0.02, 70))
plt.ylim(-0.)
plt.show()
```
